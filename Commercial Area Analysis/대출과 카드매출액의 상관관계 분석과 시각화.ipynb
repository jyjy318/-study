```python
#API 데이터 불러오기
from geoband.API import *
import pandas as pd
GetCompasData('SBJ_2012_002', '17', '17.용인시_소상공인_매출정보.csv')
df = pd.read_csv('17.용인시_소상공인_매출정보.csv')
df
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
#201703
corr = df[['sales_est_amt_201703', 'sme_loan_y_1','sme_loan_y_2','sme_loan_y_3','sme_loan_y_4','sme_loan_y_5',
          'sme_loan_y_6']].corr()
sns.heatmap(corr, cmap = 'Blues')
plt.title('lona & 201703', fontsize=20)
plt.show()
print(corr)
```
