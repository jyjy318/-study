## 크롤러의 법률적 한계
### 서비스에 지장이 올 정도로 부담을 주어서는 안된다.
1) robots.txt 를 통해 허용 범위를 알림

2) 다양한 기법으로 크롤링이 불가능하게 막음

### 수집한 정보를 통해 금전적 이득을 취해서는 안된다.
1) 제한선이 모호함

```
import requests # 서버에 접속해 데이터를 받아옴
import json # 받아온 데이터를 json 형태로 해석
```

```
# 데이터를 수집한 웹 페이지 주소 작성
stock_code = 'A005930'
url = f'https://finance.daum.net/api/quote/{stock_code}/days?symbolCode={stock_code}&page=1&perPage=100&pagination=true'
```

```
headers = {
    'user-agent':'Mozilla',
    'referer':'https://finance.daum.net/'
}
response = requests.get(url, headers=headers) # 데이터 수집을 막는 기준을 속이는 기법
```

```
if response.status_code != requests.codes.ok:
  print("접속 실패")
else:
  print("접속 성공")
```

```
json_data = json.loads(response.text) # 받아온 데이터를 json 형태로 파싱(해석)
```

```
stock_data = json_data['data'] # 해석된 데이터 중에서 필요한 요소만 추리기
```

```
for day_data in stock_data: # 데이터를 확인해 봅시다.
  print(day_data)
```

```
import pandas as pd
```

```
df = pd.DataFrame(stock_data) # 데이터를 데이터 프레임으로 변환
```
