-교집합 연산 : 일치하는것만 뽑아낸다

```
indA & indB
```

연산 : OR연산
-합집합:둘중하나만 있어도 출력한다

```
indA | indB
```

연산 : ^ -여집합 : 일치하지 않는 값만 가져온다

```
indA ^ indB
```
