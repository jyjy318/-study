```python
#pandas를 이용한 위키페이지 인구 내용 크롤링
import pandas as pd
#html형식으로 불러오기 : html 페이지 불러와서 바로 분석 가능
url = "https://ko.wikipedia.org/wiki/%EB%8C%80%ED%95%9C%EB%AF%BC%EA%B5%AD%EC%9D%98_%EC%9D%B8%EA%B5%AC"
table = pd.read_html(url) #값을 다 테이블 형식으로 가져오게 된다. 
table
```

```python
# seaborn 을 통한 위 인구 통계 시각화
import seaborn as sns
sns.set(font = 'NanumBarunGothic') 
#그래프 사이즈 조정
import matplotlib.pyplot as plt
plt.figure(figsize = (15,4))
plt.xticks(rotation=60) # 년도 기울기 조정해서 조금 보기 편하게 조정
sns.pointplot(data=df, x = '연도 (년)', y = '추계인구(명)')
```

```python
#lineplot으로 그리기
plt.figure(figsize = (15,4))
plt.xticks(rotation=60) # 년도 기울기 조정해서 조금 보기 편하게 조정
sns.lineplot(data=df, x = '연도 (년)', y = '추계인구(명)')
```
